# Question 4

```python
# Printing Original
plt.imshow(img2)
```

```python
# Nearest Neighbor
nearest_neighbor_img = cv.resize(img2, None, fx = 2, fy = 2, interpolation = cv.INTER_NEAREST)
plt.imshow(nearest_neighbor_img)
```

```python
# Bilinear
bilinear_img = cv.resize(img2, None, fx = 2, fy = 2, interpolation = cv.INTER_LINEAR)
plt.imshow(bilinear_img)
```

```python
# Bicubic
bicubic_img = cv.resize(img2, None, fx = 2, fy = 2, interpolation = cv.INTER_CUBIC)
plt.imshow(bicubic_img)
```

### Saving the image

```python
cv.imwrite("Data/Lab2/NearestNeigh.png", nearest_neighbor_img)
```

### Summary

So after rescaling these images via three methods, I found slight difference in the O/P of the three resampled/resized images.

1. Bilinear - I noticed that the image maintined the quality and didn't show signs of pixelation, there was a slight softness in the image as compared to the original. So I think Bilinear blurs out the image a bit to prevent pixelation.

2. Bicubic - This was as good as Bilinear just a bit sharper and this there were some pixels visible, but overall good.

3. Nearest Neighbor - This was by far the worst in terms of quality, lots of pixelation but the image was sharper than the original (a tiny bit).

Overall - **Bilinear** was the best method for resampling with some blur but best image quality.
