## Numpy

```python
import numpy as np

# 2D Array size
rows = 512
columns = 512

inp = open("Data/barbara_gray.raw")
print(inp)
```

```python
# Loading the input image

img_4 = np.fromfile(inp, dtype = np.uint8, count = rows*columns)
print("Dimensions:", img_4.ndim)
print("Size:", img_4.size)
print("Shape:",img_4.shape)
```

```python
# Convert from 1D to 2D array, reshaping

total_pixels = img_4.size
img_4.shape = (total_pixels // columns, columns)

# OR
# img_f = img_4.reshape(img_4, (rows, columns))
```

```python
print("New Dimensions:", img_4.ndim)
print("New Size:", img_4.size)
print("New Shape:", img_4.shape)
```

```python
# Printing the 2D array

print(img_4)
```

```python
# Saving the file into a new format

img_4.astype('int8').tofile("Data/file02.raw")
inp.close()
```
