### Dict of series

```python
my_dict = { 'one' : pd.Series([random.random(), random.random(), random.random(), random.random(), random.random()], 
                              index =['a', 'b', 'c', 'd', 'e']), 
            'two' : pd.Series([random.random(), random.random(), random.random(), random.random(), random.random()], 
                              index =['a', 'b', 'c', 'd', 'e'])} 

df = pd.DataFrame(my_dict) 
df 
```

## Question 3. Working with series

```python
# Making the series
pd1 = pd.Series(np.random.randint(0, 100, size = (100,)) )
pd2 = pd.Series(np.random.randint(0, 100, size = (100,)) )
```

```python
pd1
```

```python
pd2
```

```python
# Addition
add_pd = pd1.add(pd2)
add_pd
```

```python
sub_pd = pd1.sub(pd2)
sub_pd
```

```python
mul_pd = pd1.mul(pd2)
mul_pd
```

```python
div_pd = pd1.div(pd2)
div_pd
```

```python
my_dict4 = {'number 1' : pd1,
            'number 2' : pd2, 
            'add' : add_pd,
            'sub' : sub_pd,
            'mult' : mul_pd,
            'div' : div_pd }

df5 = pd.DataFrame(my_dict4, columns = ['number 1', 'number 2',
                              'add', 'sub', 'mult', 'div'])
```
