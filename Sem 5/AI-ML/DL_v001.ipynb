**Image**

```
import cv2 as cv
path_img = "/content/drive/My Drive/Projects/Paintings/Rem/Wallpaper_BG.png"

img = cv.imread(path_img)
img1 = cv.imread(path_img, 0)
```

```
from google.colab.patches import cv2_imshow

scale_percent = 10

#Resizing the image
resize_width = int(img.shape[0] * scale_percent/100)
resize_height = int(img.shape[1] * scale_percent/100)

img_r = cv.resize(img, (resize_height, resize_width))
cv2_imshow(img_r)
```

```
#Resizing the image
resize_width = int(img1.shape[0] * scale_percent/100)
resize_height = int(img1.shape[1] * scale_percent/100)

img1_r = cv.resize(img1, (resize_height, resize_width))
cv2_imshow(img1_r)
```

```
print(img_r.size)
print(img1_r.size)
```

```
# Printing the number of channels
print(img_r.ndim)
print(img1_r.ndim)
```

```
print("R", img_r.shape[0])
print("G", img_r.shape[1])
print("B", img_r.shape[2])
```
