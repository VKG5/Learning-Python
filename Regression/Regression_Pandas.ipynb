```python
X=file[file.columns[0:10]]
Y=file[file.columns[10]]
```

```python
XT=X.transpose()
YT=Y.transpose()
```

```python
X=X.to_numpy()
Y=Y.to_numpy()
XT=XT.to_numpy()
YT=YT.to_numpy()
```

```python
a=XT.dot(X)
b=XT.dot(Y)
```

```python
c=np.linalg.inv(a)
theta=c.dot(b)
```

```python
print(theta)
```

```python
Y_DASH=(theta.transpose()).dot(X.transpose())
```

```python
Y_DASH
```

```python
Y
```

```python
r=(Y-Y_DASH)**2
```

```python
s=0
t=0
for i in r:
    s=s+i
    t=t+1
print(t)
print("Mean Square Error = ",s/t,"%")
```
