### Processing the data

```python
# n = The number of words that will be processed
n = 1000
```

### Clue 1
#### Serially

```python
# Making a dictionary
dict_data = dict()
```

```python
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
for i in data[:n]:
    sum = 0
    for j in data[:n]:
        prob = similar(i, j)
#         print(prob)
        sum += prob*100
    dict_data[i] = sum
```

```python
# Checking the max key value pair
max(dict_data.values())
```

```python
sorted_dict = sorted(dict_data.items(), key=lambda x: x[1], reverse=True)
```

```python
sorted_dict
```

```python
clu1_sorted = []

for i in sorted_dict:
    clu1_sorted.append(i[0])
```

```python
clu1_sorted
```

### Clue 2

#### Serial

```python
def vowel(str):
    count = 0
    str = str.lower();  
    for i in range(0,len(str)):   
      if str[i] in ('a',"e","i","o","u"):  
          count = count + 1;  
      else:
        continue  
    return count
```

```python
clu2_dict = {}
for i in data[:n]:
  clu2_dict[i] = vowel(i)

clu2_dict_sorted = sorted(clu2_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
clu2_dict_sorted
```

```python
clu2_sorted = []

for i in clu2_dict_sorted:
    clu2_sorted.append(i[0])
```

```python
clu2_sorted
```

### Clue 3

#### Serial

```python
clu3_dict = {}
for i in data[:n]:
  clu3_dict[i] = len(i)

clu3_dict_sorted = sorted(clu3_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
clu3_dict_sorted
```

```python
clu3_sorted = []

for i in clu2_dict_sorted:
    clu3_sorted.append(i[0])
    
clu3_sorted
```

### Finding the password

```python
index_new = 0
index_list = []
for i in clu1_sorted:
    index_new += ((clu1_sorted.index(i) + clu2_sorted.index(i) + clu3_sorted.index(i)) * 0.33) / 0.99
    index_list.append(index_new)
    
top_100 = [x for _, x in sorted(zip(index_list, data[:n]))]
```
